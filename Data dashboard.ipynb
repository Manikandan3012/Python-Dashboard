```python
import pandas as pd
import dash
from dash import dcc
from dash import html
import plotly.express as px
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data.csv')
df.head(5)
```

```python
app = dash.Dash(__name__)

app.layout = html.Div(children=[
   html.H1(children=' Cause of Diseases Dashboard'),
   html.Div(children='''
       Analyzing the Cause of Suicide using Plotly Dash
   '''),
   dcc.Graph(
       id='histogram',
       figure={
           'data': [
               {'x': df['Cause of Suicide'], 'type': 'histogram', 'name': 'Cause of Suicide'},
           ],
           'layout': {
               'title': 'histogram-Cause of Suicide'
           }
       }
   ),
   dcc.Graph(
       id='boxplot',
       figure={
           'data': [
               {'y': df['Gender'], 'type': 'box', 'name': 'Gender'},
           ],
           'layout': {
               'title': 'Gender wise Analysis-box plot'
           }
       }
   ),
   dcc.Graph(
       id='scatterplot',
       figure={
           'data': [
               {'x': df['Cause of Suicide'], 'y': df['Tamil Nadu'], 'mode': 'Rajasthan', 'name': 'Cause of Suicide vs Tamil Nadu'},
           ],
           'layout': {
               'title': 'Cause of Suicide vs Tamil Nadu'
           }
       }
   ),
   dcc.Graph(
       id='barchart',
       figure={
           'data': [
               {'x': df['Tamil Nadu'], 'type': 'bar', 'name': 'Tamil Nadu'},
           ],
           'layout': {
               'title': 'Tamil Nadu cases'
           }
       }
   ),
   dcc.Graph(
       id='mosaicplot',
       figure=px.histogram(df, x="Cause of Suicide", color="Gender", barmode="overlay", nbins=10, width=800, height=500)
   )
    ])

```

```python
if __name__ == '__main__':
    app.run_server(port=4050)
```
